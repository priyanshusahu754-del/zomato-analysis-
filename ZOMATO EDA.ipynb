```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Zomato-data- (1).csv")
df.head(10)
```

```python
df['rate'] = df['rate'].astype(str)            # humari rating jo thi wo clean nahi thi to usko clean karne ke liye pahle saari type str krli
df['rate'] = df['rate'].str.replace('/5','')   # uske baad usme se str value /5 ko "" replce kardiya  iske baad  
df['rate'] = df['rate'].astype(float)          # str value ko frr se float me change kardiya abb humara vote column clean hai 
df.head(10)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.sort_values(by="votes", ascending=False)   # sorting acc to votes in  descending 
```

```python
df.sort_values(by="rate", ascending=False)
```

```python
top_votes = df.sort_values(by='votes', ascending=False)


top_votes[['name','votes']].head(10)
```

```python
top10 = df.sort_values(by='votes', ascending=False).head(10)


plt.figure(figsize=(10,6))
sns.barplot( x='votes', y='name', data=top10 )    # top 10 restaurant performane in vote 


plt.title("Top Restaurants by Votes")
plt.show()
```

```python
top10 = df.sort_values(by='votes', ascending=False).head(10)


plt.figure(figsize=(10,6))

sns.barplot(x='votes', y='name', data=top10)

plt.grid(True)


plt.title("Top Restaurants by Votes")

plt.show()
```

```python
top_rating = df.sort_values(by='rate', ascending=False)

top_rating[['name','rate']].head(10)
```

```python
plt.figure(figsize=(10,6))

sns.barplot(x='rate', y='name', data=top_rating.head(20)) # top 20 restaurant performane in rate 
plt.grid(True)

plt.title("Top Rated Restaurants")
plt.show()
```

# online order analysis

```python
df.groupby('online_order')['rate'].mean()  # mean of online orders rate 
```

```python
sns.boxplot(x='online_order', y='rate', data=df)

plt.title("Online Order vs Rating")        # graph of online orders rating 
plt.show()
```

# table booking analysis

```python
df.groupby('book_table')['rate'].mean() # mean of table 
```

```python
sns.boxplot(x='book_table', y='rate', data=df)

plt.title("Table Booking vs Rating")
plt.show()
```

# restaurant type analysis

```python
df.groupby('listed_in(type)')['rate'].mean().sort_values(ascending=False)
```

```python
type_rating = df.groupby('listed_in(type)')['rate'].mean().reset_index()

plt.figure(figsize=(10,6))

sns.barplot(x='listed_in(type)', y='rate', data=type_rating)

plt.title("Average Rating by Restaurant Type")

plt.show()
```

# cost VS rating analysis 

```python
df.groupby('approx_cost(for two people)')['rate'].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,6))

sns.barplot(x='approx_cost(for two people)', y='rate', data=df)

plt.title("Cost vs Rating")

plt.show()
```

# best performing restauurant 

```python
df['performance_score'] = df['rate'] * df['votes']
```

```python
best_restaurants = df.sort_values(by='performance_score', ascending=False)

best_restaurants[['name','rate','votes','performance_score']].head(10)
```

```python
top_perf = df.sort_values(by='performance_score', ascending=False).head(10)

plt.figure(figsize=(10,6))

sns.barplot(x='performance_score', y='name', data=top_perf)

plt.title("Top Performing Restaurants")

plt.show()
```

### In this analysis, we observed the following insights:

- The top 10 restaurants that perform the best were identified based on ratings and number of votes.

- Restaurants that offer online ordering tend to perform better.

- Restaurants with (buffet/café) types generally have higher ratings, which indicates better performance.

- A higher cost does not guarantee a higher rating; some restaurants with lower cost also have high ratings.
